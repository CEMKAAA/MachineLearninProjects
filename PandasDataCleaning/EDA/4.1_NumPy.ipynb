## Neden Matrisleri Kullanıyoruz ?  <a id='5'></a>

Veri biliminin çoğu alanında matrislerle çalışacağız, peki neden matrislere ihtiyacımız var? Örneğin, rakamların olduğu görsellerden onları tespit etmesini isteyeceğimiz bir derin öğrenme modeli geliştirecek olalım. Bunun için bilgisayara çok sayıda rakamların olduğu resimler verip resimlerde yazan rakamlarla birlikte eğitiyoruz. Bilgisayar bu resimlerde hangi rakam olduğunu nasıl bilebilir ki? Bilemez.

Genişliği ve uzunluğu 28 pikselden oluşan bir resim görüyorsunuz, yani toplamda 28 x 28 = 784 adet pikselcikten oluşuyor.

Soldaki resmi gördüğünde 8 olduğunu anlayabilmemiz bizim için kolay, ancak bilgisayara bunu anlatmanın yolu sayısal olarak ifade etmekten geçiyor. Aşağıda siyah beyaz 28x28 bir resmin matrise dönüştürülmüş halini görüyorsunuz. 0 ile 255 arasında değerler almış, ne kadar beyazsa 255'e o kadar yakın. Ne kadar siyahsa 0'a o kadar yakın.

Siyah beyaz bir resim olduğu için 0 ile 255 arasındaki tek bir değer alıyor her piksel. Eğer renkli olsaydı (Red Green Blue | RGB) 3 katman olacaktı ve 3 adet sayısal değer tutulacaktı her bir piksel için. Detaylarına inmeye şimdilik gerek yok, matrisi neden kullandığımızın en somut cevabını vermek için aşağıdaki görsel yeterli olacaktır.

## NumPy Array Özellikleri  <a id='7'></a>

* **ndim**: boyut sayısı

* **shape**: boyut bilgisi

* **size**: toplam eleman sayısı

* **dtype**: array veri tipi

```python
a = np.random.randint(0,10, size = (3,6))

print(a)
```

```python
print("Boyutu ", a.ndim)  # vektör olduğu için 1 boyutludur.
print("Şekli", a.shape) # shape, matrisin kaç satır ve sütundan oluştuğunu belirtir.
print("Eleman sayısı", a.size)  # eleman sayısını verir.
print("Değişken tipi ", a.dtype) # matrisin değişken tipini verir.
```

### Matrisleri Birleştirmek  <a id='10'></a>

```python
a = np.array([[1, 2, 3], 
              [4, 5, 6]])

b = np.array([[20, 30, 40], 
              [50, 60, 70]])

print(a)
print("-" * 15)                                           # 15 kere kısa çizgi yazdırdım matrislerin okunabilirliği için.
print(b)
```

```python
np.concatenate([a,b])                                                   # üst üste bindirdi matrisleri, çünkü satır ve sütun sayısı buna müsaitti. 
```

Peki ya yan yana birleştirmesini istersek?

```python
np.concatenate([a,b], axis = 1)
```

Demek ki varsayılan olarak "axis = 0" dı ve birleştirme ekseni olarak üst üste seçiliydi. Zaten Shift+TAB yaptığımızda bunu keşfedecektik.

Peki farklı boyutta matrisleri birleştirebilir miyiz?

```python
a = np.array([1, 2, 3])  # 1 satır 3 sütunluk

b = np.array([[9, 8, 7],  # 2 satır 3 sütunluk
              [6, 5, 4]])
```

```python
np.vstack([a,b])  # stack birleştirme anlamına gelir, v ifadesi vertical yani dikey anlamında kullanılmıştır.
```

Çünkü yukarıdaki birleştirmeyi ancak dikey şekilde yapabilirdik, sütun sayıları eşit olduğu için. İsterseniz horizontal(yatay) birleştirmeyi temsil eden hstack fonksiyonu ile deneyelim, hata alacağız. Çünkü satır sayıları farklıdır, yatay birleşmez.

```python
# np.hstack([a,b])  dolayısıyla bu kod patlayacaktır.
```

Ama yine de hstack() yapabileceğimiz bir matris oluşturalım.

## NumPy Elemanlarını Sıralama   <a id='11'></a>

```python
a = np.array([2, 1, 4, 99, 29,3, 5])
b = np.sort(a)
b
```

Amacımız büyükten küçüğe sıralamaksa diziyi tersine çevirebiliriz.

```python
# Listelerde dilimleme işleminde anlattığımız yöntemler Numpy'de de mevcut. 
# Burada yapılan işlem tüm diziyi göster, ancak tersten göster demektir

b[::-1]
```

Biz sıralamak istiyoruz ancak sıralanmamış halinin indislerini de bilmek istiyoruz, nasıl olur?

```python
indisler = np.argsort(a)
indisler
```
