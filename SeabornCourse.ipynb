                                            Facetgrid Oluşturma

Yanlış örnek aşağıda facetgridin nasıl kullanılmamasının gerektiğinin bir örneği var

```python
FG = sns.FacetGrid(tips, col='day', row='sex', x='total_bill', y='tip', hue='sex')
```

                                         Burada da facetgridi nasıl kullanman gerektiğin gösteriliyor

```python
FG = sns.FacetGrid(tips, col='day', row='sex')
FG.map_dataframe(sns.histplot, 'tip')
```

Başka bir örnek

```python
FG = sns.FacetGrid(tips, col='day', row='sex')
FG.map_dataframe(sns.scatterplot, x='tip', y='total_bill')
```
