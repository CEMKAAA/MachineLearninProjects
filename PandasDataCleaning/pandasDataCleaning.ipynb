Her ülkeye özel olcak şekilde en yüksek inns score`yi bul

işte bunun gibi problemlerde groupby metodunu kullanacaksın

Haydi hızlıca bir groupby ne ne değil anlayalım

```python
import pandas as pd

# Örnek Veri Seti
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [10, 15, 12, 17, 11, 16, 13, 18]
}

# Create a DataFrame from the sample data
df_orn = pd.DataFrame(data)

print("Original DataFrame:")
print(df_orn)
```

Şimdi bu elimizdeki veri setine groupby uygularsak nasıl bir şey elde edeceğiz ona bakalım

```python
# Group the DataFrame by 'Category'
grouped_df = df_orn.groupby('Category')

# Iterate over the groups
for group_name, group_data in grouped_df:
    print(f"Group: {group_name}")
    print(group_data)
    print()
```

Umarım bu görsel anlatım anlamana yardımcı olmuştur

```python
grouped_df
```

pandas dataframelerin ne olduğunu biliyorsun her bir keywordun 1d arrayler(list)`leri tutuğu bir dictionary

dolayısıyla ne zaman bir dataframeyi groupby yaparsan bu dictinarylerin tutulduğu bir arrayi elde edersin bunları görüntüleme işlemide yukarıdaki gibi for döngüsü ile yapılabilir

```python
df.groupby('Country')['Highest_Inns_Score'].max()
```

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('HighInnCountry').reset_index().sort_values('HighInnCountry', ascending=False)
```
