```python
def estimate_charges(age, w ,b):
    return w*age + b
```

```python
w = 50
b = 100
estimate_charges(30,w,b)  # tahminlerimiz Ã§ok kÃ¶tÃ¼ bu linear regression olmaya Ã§ok uzak olayÄ± biraz ilerletelim
```

```python
plt.plot(ages, estimated_charges)
plt.xlabel('Age')
plt.ylabel('estimated_charges')
```

```python
target = non_smoker_cf.charges
plt.plot(ages, estimated_charges, 'r', alpha=0.9)
plt.scatter(ages, target, s=8, alpha=0.8)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend(['Estimate', 'Actual'])
```

                                                LOSS COST FUNCTION(GENEL SAPMA HESAPLAMAK)

```python
def rmse(targets, predictions) :
    return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
rmse(target, estimated_charges)
```

```python
def try_parameters(w,b) :
    estimated_charges = estimate_charges(ages, w, b)
    plt.plot(ages, estimated_charges, 'r', alpha=0.9) # r stands for red color and alpha is tranparency if we make it 0 than the diagram will be completelly unvisible
    plt.scatter(ages, target, s=8, alpha=0.8)
    plt.xlabel('Age')
    plt.ylabel('charges')
    plt.legend(['Estimate', 'Actual'])
    
    loss = rmse(target, estimated_charges)
    print("Standart Sapma: ", loss)
```

```python
try_parameters(330, -5100)
```

                                            STANDARTIZATION(COLUMNLARI STANDARTLAMA)

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = ['age', 'bmi', 'children']
scaler = StandardScaler()
scaler.fit(medical_cf[numeric_cols])
```

```python
scaler.mean_ # HERE IS THE MEAN VALUES FOR EACH COLUMN AGE, BMI, CHILDREN
```

```python
scaler.var_ #STANDART SAPMANIN KARESÄ°
```

```python
scaler.scale_  #STANDART SAPMA
```

```python
scaled_inputs = scaler.transform(medical_cf[numeric_cols]) #STANDARTLAMA Ä°ÅžLEMÄ°: Bu iÅŸlem tÃ¼m columnlar iÃ§in uygulanÄ±r ve standartlama iÅŸleminde her bir hÃ¼credeki deÄŸerden ortalama deÄŸeri(aritmatik ort.) Ã§Ä±kartÄ±rsÄ±n daha sonra standart sapmaya bÃ¶lersin bÃ¶ylelikle verileri standartlarsÄ±n
scaled_inputs
```

```python
cat_cols = ['smoker_code', 'sex_code', 'northeast', 'northwest', 'southeast', 'southwest']
categorical_data = medical_cf[cat_cols].values
categorical_data
```

```python
inputs = np.concatenate((scaled_inputs, categorical_data), axis=1) # Elimizdeki iki arrayi biri standartlanmÄ±ÅŸ datalardan diÄŸeri categori datalarÄ±ndan oluÅŸan iki arrayi columnlardan birleÅŸtiriyoruz ve tek bir numpy arrayi elde ediyoruz
targets = medical_cf.charges
model = LinearRegression().fit(inputs, targets)
predictions = model.predict(inputs)
standart_sapma = rmse(targets, predictions)
print("Standart Sapma: ", standart_sapma)
predictions
```
